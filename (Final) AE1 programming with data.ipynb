# Abstract 
Following the preprocessing stage in which the texts were cleaned by being stripped of all punctuation and made lower case, the first model used the .count function to ascertain the quantity of exact colour terms. This was compared to the quantity of substrings for the same colour terms in both texts. Due to a suspiciously high disparity between the two sets of figures, a substring finding functions was build. When it became evident that most of the substrings fell in one of five categories which were semantically similar to the colour terms, a new model was build with the .split function which included such substrings. The data produced from this model was used for the final analysis regarding the proportion of colour terms in both books. Finally the model was tested with a text file containing exclusively colour terms. 

# Preprocessing and Inputting terms
After cleaning the data we are left with the following variables:
* 'opt_clean' for Newton's Opticks
* 'goe_clean' for Goethe's Theory of Colours

These are used throughout.

```python
optics = open("optics.txt","r")
opt_str = optics.read().lower()
goethe = open("goethe's.txt", "r")
goe_str = goethe.read().lower()

def punct_strip(book):
    punct = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    book_clean = ""
    for i in book:
        if i not in punct:
            book_clean += i
    return book_clean

opt_clean = punct_strip(opt_str)
goe_clean = punct_strip(goe_str)
```

# Experiment 1
In my first experiment, I try to avoid 'substrings' in my count altogether. This is done by surrounding the 'i' value in line 4 with spaces so that only exact words and thus exact colour terms can be counted.


## Exact colour term count

```python
col_list = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'purple', 'violet', 'black', 'white']

def colour_counter(col_list, book):
    col_counts = [book.count(" " + i + " ") for i in col_list] 
    return col_counts

col_counts_opt = colour_counter(col_list, opt_clean)
for i in range(len(col_counts_opt)):
    print("There are " + str(col_counts_opt[i]) + " occurrences of the colour term " + "'" + col_list[i] + "'" + " in Newton's Opticks" )
```

```python
col_counts_goe = colour_counter(col_list, goe_clean)
for i in range(len(col_counts_goe)):
    print("There are " + str(col_counts_goe[i]) + " occurrences of the colour term " + "'" + col_list[i] + "'" + " in Goethe's Theory of Colours " )
```

## Substring and colour term sum results 
Here I make one alteration to the function 'colour_counter' which is removing the spaces adjacent to the 'i' value. The result of this is that it will count the number of colour terms and substrings returning their sum.

```python
def colour_counter_substr(col_list, book):
    col_counts_substr = [book.count(i) for i in col_list]
    return col_counts_substr

col_counts_opt_substr = colour_counter_substr(col_list, opt_clean)
for i in range(len(col_counts_opt_substr)):
    print("There are " + str(col_counts_opt_substr[i]) + " occurrences of the term or substring " + "'" + col_list[i] + "'" + " in Newton's Opticks" )
```

# Investigation into quantity of Substrings
## Substring finding function
The function 'exception finder' has three components to it:
* It counts the number of exact colour terms via a diffent method as a check
* It does the same with the sum of colour terms and substrings.
* It provides every occurances of a substring for a given colour for a given book. 
* The user is free to experiment with this function by calling it with different colours or books on line 13.

```python
def exception_finder(book, term):
    tokens = book.split()
    exceptions = [i for i in tokens if term in i and term != i]
    pure_count = 0
    sub_count = 0
    for i in tokens:
        if i == term:
            pure_count += 1
        if term in i:
            sub_count += 1
    return pure_count, sub_count, exceptions

pure_count, sub_count, exceptions = exception_finder(opt_clean, "violet")
print("exact count: ", pure_count, ' \nexact + sub count: ', sub_count, ' \nExceptions: ', exceptions)
```

## Analysis of results from investigation 
* This new approach utilising the .split method has slighlty increased counts for the exact colours terms in both books.
* The same approach, however, results in the same amount of exact terms + substrings. This suggests that this method is more inclusive or less stringent in its definition of an exact colour terms to substrings rather than simply counting more words. 
* Since our problem with the model in Experiment 1 was that it seemed to be too stringent, the .split method used here seems to preferable.
* The vast amount of substrings is mainly attributable to substrings of the form, colour term + 'making', 's', 'ness', 'ish' or 'er'. Examples would thus be 'whitemaking' 'whites' 'whiteness' 'whiteish' and 'whiter'.

# Experiment 2
* We now have the option of increasing our colour terms count and consequenlty decreasing our subsring count by using the .split method as well as including specific substrings types.
* I decided this would be appropriate in the case of the above examples
* If one wished to add more substring exceptions to the colour count more elif lines can simply be added to the function 'full_count'.

```python
def full_count2(book, x=col_list):
    tokens = book.split()
    
    p_c_lst = []
    g_e_lst = []
    for i in col_list:
        pure_count = 0
        good_exceptions = 0
        for word in tokens:
            if i == word:
                pure_count += 1
            if word == i + "making":
                good_exceptions += 1
            elif word == i + "s":
                good_exceptions += 1
            elif word == i + "ness":
                good_exceptions += 1
            elif word == i + "er":
                good_exceptions += 1
            elif word == i + "ish":
                good_exceptions += 1
        p_c_lst.append(pure_count)
        g_e_lst.append(good_exceptions)
    final_count_lst = [p_c_lst[i] + g_e_lst[i] for i in range(len(g_e_lst))]
    return final_count_lst

for i in range(len(col_list)):
    print("For the colour '" + col_list[i] + "' the more extensive counts are " + str(full_count2(opt_clean)[i]) + " for Newton's Opticks and " + str(full_count2(goe_clean)[i]) + " for Goethe's Theory of Colour")
```

# Final Analysis
Here I am simply working out the percentage occurance of each colour term in the books and colour terms as a whole

```python
def percentage(book):
    word_count = len(book.split())
    col_freq = full_count2(book)
    percentages = [i / word_count * 100 for i in col_freq]
    return percentages 

        
perc_lst_opt = percentage(opt_clean)
for i in range(len(perc_lst_opt)):
    print("The proportion of the term '" + col_list[i] + "'" + " in Newton's Opticks is " + str(perc_lst_opt[i]) + "%")

print("")

total_opt = 0
for i in perc_lst_opt:
    total_opt += i

print("The total proportion of colour terms in Newton's Opticks is " + str(total_opt) + "%")
```

```python
perc_lst_goe = percentage(goe_clean)
for i in range(len(perc_lst_goe)):
    print("The proportion of the term '" + col_list[i] + "'" + " in Goethe's Theory of Colours is " + str(perc_lst_goe[i]) + "%")

print("")

total_goe = 0
for i in perc_lst_goe:
    total_goe += i

print("The total proportion of colour terms in Goethe's Theory of Colours is " + str(total_goe) + "%")
```

# Ethical Reflections 
In an age of growing disinformation fact checking algorithms (FCAs) are increasingly vital but FCAs presuppose a more fundemental type of algorithm, namely ones which extract facts about the world. These must be equally objective and inviolable if we are to build trust in FCAs, yet as this project demonstrates this is often more difficult that it seems. 'Fact' can be a more subjective than objective matter. In the case of the colour counter, only counting exact terms intuitively seemed overly precise and restrictive. Whether, however, terms such as 'white-making' or 'whiter' should be included is also not clear. They may be semantically similar but are syntactically different, the former being an adjective and the latter being a relation as opposed to nouns like the exact colour terms. 

My inital reaction was to alter my code so that the user could choose whether to include such terms or not but this would lead to a lack of uniformity in results. I believe such an approach has the potential to proliferate the problem of disinformation futher. As such, I decided that as the data scientist, it was my perogative to decide which terms ought to be included in the count however, since such decisions are somewhat arbitrary in nature, transparency is essential.
